```python
import requests
import time
import os
import json
import base64
import statistics

folder_path = "extended_records"
records = os.path.join(os.getcwd(), folder_path)

all_times = {
    "ops1": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "ops2": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "ops3": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "ops4": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "ops5": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "EVALUATE_LICENSE": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "FIND_ANNOTATION": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "FIND_METADATA": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "GET_ORCID": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "GET_THUMBNAIL": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
    "VALIDATE_SCHEMA": {"values_w_api": [], "avg_w_api": 0, "std_w_api": 0, "values_w_o_api": [], "avg_w_o_api": 0, "std_w_o_api": 0},
}
```

```python
def execute_tpm_adapter(ops_type, payload):
    # Start timing
    start_time = time.perf_counter()
    # For a GET request
    #response = requests.get(url, params=data)

    # For a POST request
    url = "http://127.0.0.1:5001/doip"
    response = requests.post(url, json=payload)
    end_time = time.perf_counter()
    elapsed_time_w_api = end_time - start_time
    all_times[ops_type]["values_w_api"].append(elapsed_time_w_api*1000)
    try:
        all_times[ops_type]["avg_w_api"] = statistics.mean(all_times[ops_type]["values_w_api"])
        all_times[ops_type]["std_w_api"] =  statistics.stdev(all_times[ops_type]["values_w_api"])
    except statistics.StatisticsError:
        all_times[ops_type]["avg_w_api"] = 0
        all_times[ops_type]["std_w_api"] = 0
    # time without operation response
    elapsed_time_w_o_api=response.json()["elapsed_time"]
    all_times[ops_type]["values_w_o_api"].append(elapsed_time_w_o_api*1000)
    try:
        all_times[ops_type]["avg_w_o_api"] = statistics.mean(all_times[ops_type]["values_w_o_api"])
        all_times[ops_type]["std_w_o_api"] =  statistics.stdev(all_times[ops_type]["values_w_o_api"])
    except statistics.StatisticsError:
        all_times[ops_type]["avg_w_o_api"] = 0
        all_times[ops_type]["std_w_o_api"] = 0
    return response.json()["response"], response
```

```python
#listing all operations available by the service
ops1 = {
    "operationId": "0.DOIP/OP.LIST_OPS",
    "targetId": "SERVICE"
}
response_message, full_response = execute_tpm_adapter("ops1", ops1)
print(json.dumps(response_message, indent=4))
```

```python
#listing all FDOs registered in the TPM instance
ops2 = {
    "operationId": "0.DOIP/Op.LIST_FDOs",
    "targetId": "SERVICE"
}
response_message, full_response  = execute_tpm_adapter("ops2", ops2)
print(json.dumps(response_message, indent=4))
```

```python
ops_pids = []
all_pids = []

#reading known FDOs from the records folder for more convenience. Can also be achieved using ops2 (typical approach) but requires record inspection to distinct FDO-Ops from FDOs
for filename in os.listdir(records):
    if ("ops" in filename) and (filename.endswith('.json')):
        operations = os.path.join(records, filename)
        
        with open(operations, 'r') as file:
            operation_record = json.load(file)
        
        ops_pids.append(operation_record["pid"])
        all_pids.append(operation_record["pid"])

    elif filename.endswith('.json'):
        non_operations = os.path.join(records, filename)
        
        with open(non_operations, 'r') as file:
            non_operation_record = json.load(file)

        all_pids.append(non_operation_record["pid"])
```

```python
# Get the information record from each FDO PID
for pid in all_pids:
    ops4 = {
    "operationId": "0.DOIP/Op.GET_FDO",
    "targetId": pid
    }
    response_message, full_response  = execute_tpm_adapter("ops4", ops4)
    print(f"information record for {pid}", json.dumps(response_message, indent=4))
```

```python
# Get the list of supported FDO-Ops for a FDO (including other FDO-Ops)

for pid in all_pids:
    ops5 = {
    "operationId": "0.DOIP/Op.LIST_OPS",
    "targetId": pid
    }
    response_message, full_response  = execute_tpm_adapter("ops5", ops5)
    print(f"supported FDOps for {pid}", json.dumps(response_message, indent=4))
```

```python
# Get the list of FDOps for a specific FDO
for pid in all_pids:
    ops5 = {
    "operationId": "0.DOIP/Op.LIST_OPS",
    "targetId": pid
    }
    response_message, full_response  = execute_tpm_adapter("ops5", ops5)
    
    # Execute supported FDOps knowing the names to also track execution times
    for ops in response_message['available FDOps']:
        ops_id = ops["pid"]
        ops_name = ops["name"]
        # knowing the EVALUATE_LICENSE operation requires additional parameters, as described in the record
        if ops_name == "EVALUATE_LICENSE":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid,
            "attributes": {
                "21.T11148/916ca3badfa68b06870c": "open access"
                }
            }
            ops_response_message, ops_full_response  = execute_tpm_adapter("EVALUATE_LICENSE", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
        # knowing the GET_THUMBNAIL returns a png image that needs to be encoded and should be saved
        elif ops_name == "GET_THUMBNAIL":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid
            }
            ops_response_message, ops_full_response  = execute_tpm_adapter("GET_THUMBNAIL", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
            parts = pid.split('/')
            filename = parts[1]
            if ops_full_response.status_code == 200:
                # Decoding the Base64 strings back to binary data
                print("!",ops_full_response.json()["response"]["operation result"])
                decoded_response = {key: base64.b64decode(value) for key, value in ops_full_response.json()["response"]["operation result"].items()}
                for request, data in decoded_response.items():
                    with open(f'{filename}_{request}.png', 'wb') as image_file:
                        image_file.write(data)
            else:
                print(f"Error: {ops_full_response.text}")
        elif ops_name == "FIND_ANNOTATION":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid
            }
            ops_response_message, ops_full_response = execute_tpm_adapter("FIND_ANNOTATION", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
        elif ops_name == "FIND_METADATA":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid
            }
            ops_response_message, ops_full_response = execute_tpm_adapter("FIND_METADATA", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
        elif ops_name == "VALIDATE_SCHEMA":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid
            }
            ops_response_message, ops_full_response = execute_tpm_adapter("VALIDATE_SCHEMA", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
        elif ops_name == "GET_ORCID":
            ops6 = {
            "operationId": ops_id,
            "targetId": pid
            }
            ops_response_message, ops_full_response = execute_tpm_adapter("GET_ORCID", ops6)
            print(f"result for operation {ops_name} with PID {ops_id} and FDO {pid}", json.dumps(ops_response_message, indent=4))
        else:
            pass
```
